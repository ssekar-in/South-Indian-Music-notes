```python
import numpy as np
import IPython.display as ipd
from scipy.io.wavfile import write

Sa=1/1; R1f=256/243;R1s=16/15; R2f=10/9; R2s=9/8; R3f= 32/27; R3s=6/5
G1f=R2f;G1s=R2s; G2f=R3f; G2s=R3s; G3f= 5/4; G3s=81/64
M1f=4/3; M1s=27/20; M2f=45/32;M2s=729/512; Pa=3/2
D1f=128/81; D1s= 8/5; D2f= 5/3; D2s= 27/16; D3f=16/9; D3s=9/5
N1f=D2f; N1s=D2s; N2f=D3f; N2s=D3s; N3f= 15/8; N3s=243/128; HSa=2/1

def generate_sinusoid_scale(pitches=[Sa], duration=0.5, Fs=4000, amplitude_max=0.5):
    """Generate synthetic sound of scale using sinusoids

    Notebook: C5/C5S1_Scales_CircleFifth.ipynb

    Args:
        pitches (list): List of pitches (freq ratio) (Default value = [1/1])
        duration (float): Duration (seconds) (Default value = 0.5)
        Fs (scalar): Sampling rate (Default value = 4000)
        amplitude_max (float): Amplitude (Default value = 0.5)

    Returns:
        x (np.ndarray): Synthesized signal
    """
    N = int(duration * Fs)
    t = np.arange(0, N) / Fs
    
    x = []
    f=240
    om= [1,2,3,4,5,6,7,8]
    w=[1, 0.3, 0.2, 0.1, 0.1, 0.075, 0.075, 0.05]
    
    for p in pitches:
        xa=0
        n=0
        for wa in w:
            xa = xa + (w[n]*np.sin(2*np.pi*om[n]*p*f*t))
            n=n+1
        x = np.append(x,(xa))
    x = amplitude_max * x / np.max(x)
    return x

duration = 0.7
Fs = 8000
# Devagandhari
raag = 'Mayamalava gowla'
pitches = [Sa,R1s,G3s,M1f,Pa,D1s,N3f,HSa,HSa,N3f,D1s,Pa,M1f,G3s,R1s,Sa]
x = generate_sinusoid_scale(pitches=pitches, duration=duration, Fs=Fs)
print(raag, flush=True)
ipd.display(ipd.Audio(data=x, rate=Fs))
write('raag.wav', Fs, x)
# rename raag.wav in home folder before running another raagam
```
